```python
from pyspark.sql import SparkSession

# Initialize a local Spark session
spark = SparkSession.builder \
    .appName("VSCode-Spark-Local") \
    .master("local[*]") \
    .getOrCreate()

# Create dummy data to test
data = [("Alice", 34), ("Bob", 45), ("Catherine", 29)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, schema=columns)
df.show()
```

```python

# Read JSON data from a URL and create a DataFrame
# Source- https://github.com/apache/spark/tree/master

import urllib

url = "https://raw.githubusercontent.com/apache/spark/master/examples/src/main/resources/people.json"
data = urllib.request.urlopen(url).read().decode("utf-8")

rdd = spark.sparkContext.parallelize(data.splitlines())
df = spark.read.json(rdd)
df.show()
```
